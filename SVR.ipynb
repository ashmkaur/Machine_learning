```python
#Polynomial Regression

#Classification

#SUPERVISED LEARNING 

import numpy as py
import matplotlib.pyplot as plt
import pandas as pd

dataset= pd.read_csv('Poly_dataset.csv')
X=dataset.iloc[:,:-1].values
Y=dataset.iloc[:,1].values

#Standard and fir the data for better prediction
from sklearn.preprocessing import StandardScaler
sc_X=StandardScaler()
X= sc_X.fit_transform(X)
sc_Y=StandardScaler()
Y= sc_Y.fit_transform(py.reshape(Y,(10,1)))

# Polynomial regression
from sklearn.svm import SVR
lin_reg= SVR(kernel='rbf')
lin_reg.fit(X,Y)

# Linear

plt.scatter(X,Y,color='red')
plt.plot(X,lin_reg.predict(X),color='black')
plt.title("Level VS Salary")
plt.xlabel("Level")
plt.ylabel("Salary")
plt.show()


# which quite obvious regarding interval in which 6.5 lies 

lin_reg.predict(py.reshape(6.5,(1,1)))
predict = sc_Y.inverse_transform(lin_reg.predict(sc_X.transform(py.array([[6.5]]))).reshape(-1, 1))
# lin_reg_2.predict(poly_reg.fit_transform([[6.5]]))
print(predict)
```
