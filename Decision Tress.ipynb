```python
#Polynomial Regression

#Classification

#SUPERVISED LEARNING 

import numpy as py
import matplotlib.pyplot as plt
import pandas as pd

dataset= pd.read_csv('Poly_dataset.csv')
X=dataset.iloc[:,:-1].values
Y=dataset.iloc[:,1].values


# Polynomial regression
from sklearn.tree import DecisionTreeRegressor
lin_reg= DecisionTreeRegressor(random_state=0)
lin_reg.fit(X,Y)

plt.scatter(X,Y,color='red')
plt.plot(X,lin_reg.predict(X),color='black')
plt.title("Level VS Salary")
plt.xlabel("Level")
plt.ylabel("Salary")
plt.show()

# Regressor
x_grid= py.arange(min(X),max(X),0.1)
x_grid= x_grid.reshape(len(x_grid),1)
plt.scatter(X,Y,color='red')
plt.plot(x_grid,lin_reg.predict(x_grid),color='black')
plt.title("Level VS Salary")
plt.xlabel("Level")
plt.ylabel("Salary")
plt.show()


# which quite obvious regarding interval in which 6.5 lies 

lin_reg.predict(py.reshape(9.5,(1,1)))
```
