```python
#Polynomial Regression

#Classification

#SUPERVISED LEARNING 

import numpy as py
import matplotlib.pyplot as plt
import pandas as pd

dataset= pd.read_csv('Poly_dataset.csv')
X=dataset.iloc[:,:-1].values
Y=dataset.iloc[:,1].values

X
# from sklearn.model_selection import train_test_split
# X_train,X_test,y_train,y_test= train_test_split(X,Y,test_size=0.3,random_state=0)

# Polynomial regression
from sklearn.linear_model import LinearRegression
lin_reg= LinearRegression()
lin_reg.fit(X,Y)

from sklearn.preprocessing import PolynomialFeatures
poly_reg= PolynomialFeatures(degree=2)
X_poly=poly_reg.fit_transform(X)
poly_reg.fit(X_poly,Y)
lin_reg_2=LinearRegression()
# This is fit on linear regression because it is linear wrt b cofficients that's ww transformed it to squares then plotted with linear regression only.

lin_reg_2.fit(X_poly,Y)

# Linear

plt.scatter(X,Y,color='red')
plt.plot(X,lin_reg.predict(X),color='black')
plt.title("Level VS Salary")
plt.xlabel("Level")
plt.ylabel("Salary")
plt.show()

# Polynomial

plt.scatter(X,Y,color='red')
plt.plot(X,lin_reg_2.predict(poly_reg.fit_transform(X)),color='black')
plt.title("Salary VS Experience")
plt.xlabel("Years")
plt.ylabel("Salary")
plt.show()

# which quite obvious regarding interval in which 6.5 lies 

lin_reg.predict([[6.5]])
lin_reg_2.predict(poly_reg.fit_transform([[6.5]]))



```
